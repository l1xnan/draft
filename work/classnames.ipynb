```python
book = xw.Book('classnames.xlsx')
```

```python
sht = book.sheets['汇总']
```

```python
rng = sht.range('A1').current_region.options(numbers=int).value[1:]
```

```python
class_id = [
    '悉尼一班',
     '悉尼二班',
     '哥伦比亚一班',
     '哥伦比亚二班',
     '哥伦比亚三班',
     '普林斯顿一班',
     '普林斯顿二班',
     '普林斯顿三班',
     '多伦多一班',
     '多伦多二班',
     '斯坦福一班',
     '斯坦福二班',
     '牛津一班',
     '剑桥一班',
     '剑桥二班',
     '耶鲁一班',
     '耶鲁二班',
     '哈佛一班',
]
```

```python
rng = sht.range('B2').expand('table').options(numbers=int).value
rng
```

```python
D = itertoolz.groupby(lambda r: r[1], rng)
```

```python
workbook = xlsxwriter.Workbook('名单.xlsx')
worksheet = workbook.add_worksheet('名单')
title = ['序号', '班级', '姓名', '性别', '是否寄宿', '房间号', '备注']
row, col = 0, 0
pagebreaks = []
for key in class_id:
    cls = [title] + D[key]
    for i in range(len(cls)):
        for j in range(len(cls[0])):
#             print(cls[i])
            try:
                worksheet.write(row, j, cls[i][j])
            except IndexError:
                pass
        row = row + 1
#     worksheet.set_h_pagebreaks([row])
    pagebreaks.append(row)
worksheet.set_h_pagebreaks(pagebreaks)
workbook.close()
print('OK')
```
